# **Data training and preprocessing**

we will try two different clusters (km,hc) 

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_scaled=scaler.fit(df)
df_scaled=df_scaled.transform(df)
```

# **K_means**

```python
from sklearn.cluster import KMeans

kmeans = KMeans(3)
kmeans
```

```python
df.head()
```

```python
K_pred = kmeans.fit_predict(df_scaled)
K_pred
```

```python
df['K_means_Cluster'] = K_pred
df.head()
```

```python
df1 = df[df['K_means_Cluster']==0]
df2 = df[df['K_means_Cluster']==1]
df3 = df[df['K_means_Cluster']==2]


plt.figure(figsize=(10, 10))

sns.scatterplot(x='Alcohol', y='Color_Intensity', data=df1, color='green',label='c1')
sns.scatterplot(x='Alcohol', y='Color_Intensity', data=df2, color='red',label='c2')
sns.scatterplot(x='Alcohol', y='Color_Intensity', data=df3, color='blue',label='c3')


plt.xlabel('Alcohol')
plt.ylabel('Color_Intensity')

plt.show()
```

```python
df1 = df[df['K_means_Cluster']==0]
df2 = df[df['K_means_Cluster']==1]
df3 = df[df['K_means_Cluster']==2]


plt.figure(figsize=(10, 10))

sns.scatterplot(x='Alcohol', y='Proline', data=df1, color='green',label='c1')
sns.scatterplot(x='Alcohol', y='Proline', data=df2, color='red',label='c2')
sns.scatterplot(x='Alcohol', y='Proline', data=df3, color='blue',label='c3')


plt.xlabel('Alcohol')
plt.ylabel('Proline')

plt.show()
```

```python
kmeans.cluster_centers_
```

```python
k_rng = range(1,10)
sse = []

for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df[['Alcohol', 'Color_Intensity']])
    sse.append(km.inertia_)
```

```python
fig = plt.figure(figsize=(8,6))

plt.xlabel('K')
plt.ylabel('Sum of Squred Error')
plt.plot(k_rng,sse)
```

as the plot shows we have 3 clusters

```python
df1 = df[df['Hierarchical_Cluster']==0]
df2 = df[df['Hierarchical_Cluster']==1]
df3 = df[df['Hierarchical_Cluster']==2]

plt.figure(figsize=(8, 6))

sns.scatterplot(x='Alcohol', y='Proline', data=df1, color='green',label='c1')
sns.scatterplot(x='Alcohol', y='Proline', data=df2, color='red',label='c2')
sns.scatterplot(x='Alcohol', y='Proline', data=df3, color='blue',label='c3')

plt.xlabel('Alcohol')
plt.ylabel('Proline')

plt.show()
```

it is prefarable to do dimensionality reduction to make the cluster more clear so we will use PCA for that 

# **PCA**

```python
from sklearn.decomposition import PCA
pca=PCA(2)
pca_data=pca.fit_transform(df_scaled)
pca_data.shape
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams["figure.dpi"] = 80
plt.scatter(x=pca_data[:,0], y=pca_data[:,1], lw=2)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA")
plt.show()
```

```python
pca_dataframe = pd.DataFrame(pca_data, columns=["PCA 1", "PCA 2"])
pca_dataframe["K_Cluster"] = K_pred
pca_dataframe["HC_Cluster"] = HC_pred
pca_dataframe.sample(10)
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams["figure.dpi"] = 80
sns.scatterplot(x="PCA 1", y="PCA 2", hue="K_Cluster", data=pca_dataframe, palette=['green','red','blue'], s=100)
plt.title("K_PCA")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams["figure.dpi"] = 80
sns.scatterplot(x="PCA 1", y="PCA 2", hue="HC_Cluster", data=pca_dataframe, palette=['green','red','blue'], s=100)
plt.title("HC_PCA")
plt.show()
```

```python
df["PCA 1"] = pca_dataframe["PCA 1"]
df["PCA 2"] = pca_dataframe["PCA 2"]
df["K_PCA"] = pca_dataframe["K_Cluster"]
df["HC_PCA"] = pca_dataframe["HC_Cluster"]
df.head()
```

```python
c_type = df[['Alcohol', 'Color_Intensity','Proline', 'K_means_Cluster', 'Hierarchical_Cluster','PCA 1','PCA 2','K_PCA','HC_PCA']]
c_type.head()
```

# **Cluster visualization**

```python
K_pca = c_type['K_PCA'].value_counts()
HC_pca = c_type['HC_PCA'].value_counts()
print(HC_pca,'\n',K_pca)
```

```python
HC_PCA= pd.DataFrame({
    'HC_pca_Cluster' : HC_pca.index,
    'HC_pca' : HC_pca.values
}, columns=['HC_pca_Cluster', 'HC_pca'])

HC_PCA.set_index('HC_pca_Cluster', inplace=True)
HC_PCA
```

```python
K_PCA= pd.DataFrame({
    'K_pca_Cluster' : K_pca.index,
    'K_pca' : K_pca.values
}, columns=['K_pca_Cluster', 'K_pca'])

K_PCA.set_index('K_pca_Cluster', inplace=True)
K_PCA
```

```python
result = pd.concat([K_PCA,HC_PCA], axis=1)
result
```

```python
import plotly.graph_objs as go
```

```python
trace1 = go.Bar(
    x=result.index,
    y=result['K_pca'],
    marker_color='red',
    name='K_pca'
)

trace2 = go.Bar(
    x=result.index,
    y=result['HC_pca'],
    marker_color='black',
    name='HC_pca'
)



data = [trace1, trace2]
layout = go.Layout(
    barmode='stack',
    title=" Number of values that belong to each cluster",
    xaxis={
        'title':"Cluster",
    },
    yaxis={
        'title':"Cluster NUmber",
    }
)
figure=go.Figure(data=data,layout=layout)
figure
```
